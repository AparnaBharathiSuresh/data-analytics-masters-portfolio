```python
import matplotlib.pyplot as plt
import numpy as np
num_of_rolls = [100, 1000, 5000, 10000, 50000, 100000, 300000, 700000, 1000000]

for n in num_of_rolls:
    #Simulate n die rolls
    rolls = np.random.randint(1,7,size=n)
    
    #Calculate the probability of each face value
    _,counts = np.unique(rolls, return_counts= True)
    probability = counts / n
    
    #Create Bar chart
    fig, ax = plt.subplots(figsize=(5, 5))
    bar_labels = ['1', '2', '3', '4', '5', '6']
    bar_colors = ['brown', 'grey', 'pink', 'orange', 'green', 'red']
    bars = ax.bar(range(1, 7), probability, label = bar_labels, color = bar_colors )
    ax.set_xlabel('Die Face Value')
    ax.set_ylabel('Probability')
    ax.set_title(f"Frequency Distribution of Die Rolls (n={n})")
    ax.legend(loc='upper right',fontsize=6)
    plt.show()
```

```python
import matplotlib.pyplot as plt
import random

class Cards:
    #Define the shape, numbers and face cards
    Category=["Hearts","Diamonds","Clubs","Spades"]
    Number=["2","3","4","5","6","7","8","9","10","Ace","Jack","Queen","King"]
    Face=["Jack","Queen","King"]
    
    #Initialize the object
    def __init__(self,shape,value):
        self.shape=shape
        self.value=value
    
    #Method to check Event A
    def is_Diamond(self):
        if self.shape== "Diamonds":
            return True
        else:
            return False
        
     #Method to check Event B
    def is_Non_FaceCard(self):
        if self.value not in Cards.Face:
            return True
        else:
            return False
     
     #Method to check Event C
    def is_Divisible_By_4(self):
        if self.value.isdigit():
            return int(self.value) % 4 == 0
        else:
            return False
        
#Define the number of trials
No_of_trials = [100, 5000, 10000, 50000, 100000, 300000, 700000, 1000000]

#Create empty array to store the probabilities in each trial
Prob_A_Array, Prob_B_Array, Prob_C_Array, Prob_AB_Array, Prob_AC_Array, Prob_BC_Array,Prob_ABC_Array, Prob_unionABC_Array=[],[],[],[],[],[],[],[]

for n in No_of_trials:
    #Initialize the count variables
    Count_A,Count_B,Count_C,Count_AB,Count_AC,Count_BC,Count_ABC=0,0,0,0,0,0,0
    
    #Generate a random card
    Obj = [Cards(random.choice(Cards.Category),random.choice(Cards.Number)) for _ in range(n)]
    for i in range(n):
        #Check if the event has occurred
        if(Obj[i].is_Diamond()):
            Count_A+=1
        
        if(Obj[i].is_Non_FaceCard()):
            Count_B+=1
        
        if(Obj[i].is_Divisible_By_4()):
            Count_C+=1
        
        if(Obj[i].is_Diamond() and Obj[i].is_Non_FaceCard()):
            Count_AB+=1
        
        if(Obj[i].is_Non_FaceCard() and Obj[i].is_Divisible_By_4()):
            Count_BC+=1
        
        if(Obj[i].is_Diamond() and Obj[i].is_Divisible_By_4()):
            Count_AC+=1
        
        if(Obj[i].is_Diamond() and Obj[i].is_Non_FaceCard() and Obj[i].is_Divisible_By_4()):
            Count_ABC+=1
    
    #Calculate the Probability of each event
    Prob_A=Count_A/n
    Prob_B=Count_B/n
    Prob_C=Count_C/n
    Prob_AB=Count_AB/n
    Prob_AC=Count_AC/n
    Prob_BC=Count_BC/n
    Prob_ABC=Count_ABC/n
    Prob_union_ABC = Prob_A + Prob_B + Prob_C - Prob_AB - Prob_AC - Prob_BC + Prob_ABC
    Prob_A_Array.append(Prob_A)
    Prob_B_Array.append(Prob_B)
    Prob_C_Array.append(Prob_C)
    Prob_AB_Array.append(Prob_AB)
    Prob_AC_Array.append(Prob_AC)
    Prob_BC_Array.append(Prob_BC)
    Prob_ABC_Array.append(Prob_ABC)
    Prob_unionABC_Array.append(Prob_union_ABC)

#Plot the graph

plt.figure(figsize=(20, 20))

# Event A
plt.subplot(3, 3, 1)  
plt.plot(No_of_trials, Prob_A_Array, label="Event A (Diamond)")
plt.xlabel("Number of Trials")
plt.ylabel("Probability")
plt.title("Event A: Probability of diamond card with Number of Trials")
plt.legend()

# Event B
plt.subplot(3, 3, 2)  
plt.plot(No_of_trials, Prob_B_Array, label="Event B (Non Face Card)")
plt.xlabel("Number of Trials")
plt.ylabel("Probability")
plt.title("Event B: Probability of non face card with Number of Trials")
plt.legend()

# Event C
plt.subplot(3, 3, 3)  
plt.plot(No_of_trials, Prob_C_Array, label="Event C (Divisible by 4)")
plt.xlabel("Number of Trials")
plt.ylabel("Probability")
plt.title("Event C: Probability of card divisible by 4 with Number of Trials")
plt.legend()

# Event AB
plt.subplot(3, 3, 4)  
plt.plot(No_of_trials, Prob_AB_Array, label="Event AB")
plt.xlabel("Number of Trials")
plt.ylabel("Probability")
plt.title("Event A INTERSECTION B")
plt.legend()

# Event AC
plt.subplot(3, 3, 5)  
plt.plot(No_of_trials, Prob_AC_Array, label="Event AC")
plt.xlabel("Number of Trials")
plt.ylabel("Probability")
plt.title("Event A INTERSECTION C")
plt.legend()

 # Event BC
plt.subplot(3, 3, 6) 
plt.plot(No_of_trials, Prob_BC_Array, label="Event BC")
plt.xlabel("Number of Trials")
plt.ylabel("Probability")
plt.title("Event B INTERSECTION C")
plt.legend()

 # Event ABC
plt.subplot(3, 3, 7) 
plt.plot(No_of_trials, Prob_unionABC_Array, label="Event Union ABC")
plt.xlabel("Number of Trials")
plt.ylabel("Probability")
plt.title("Event UnionABC")
plt.legend()


#plot all together  
plt.subplot(3, 3, 8)
plt.plot(No_of_trials, Prob_A_Array, label = "Event A(Diamond)")
plt.plot(No_of_trials, Prob_B_Array, label = "Event B(Non Face Card)")
plt.plot(No_of_trials, Prob_C_Array, label = "Event C(Divisible By 4)") 
plt.plot(No_of_trials, Prob_AB_Array, label = "Event Intersection AB") 
plt.plot(No_of_trials, Prob_AC_Array, label = "Event Intersection Ac")
plt.plot(No_of_trials, Prob_BC_Array, label = "Event Intersection BC")
plt.plot(No_of_trials, Prob_unionABC_Array, label = "Event UnionABC")
plt.xlabel("Number of Trials")
plt.ylabel("Probability")
plt.title("All Together")
plt.legend()
plt.show()
```
