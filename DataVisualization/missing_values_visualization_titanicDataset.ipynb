Q1: List three strategies for handling missing values:

Answer:

1. Mean/Median Imputation:

    -This involves replacing the missing values with mean or median value of the column. This method is suitable for      continuous or categorical data. fillna() method from the pandas are used to make these replacements.
    -Its simple and easy to implement, these are some of the benefits of the mean/median imputation.
    -The imputation is performed using the existing information from the non-missing data; hence no additional data is required.
    -Suitable for Normally distributed data.
    
2. Deletion of data:

    -This invloves removing rows or columns with missing values.
    -It is simple but can lead to loss of valuable information, especially if lots of data are missing.
    -It is generally used and beneficial when missing values are low and have no importance.
   
3. Random Sample Imputation:

    -Two subsets are created from the original data.
    -One subset with observations without missing data, and the second one with missing data. 
    -Random obervation is selected from each subsets.
    -The missing data from the previously selected observation is replaced with the existing ones from the  observation having all the data available.
    -The process is repeated until there is no more missing data.
    -It tackles both numerical and categorical data types.
    

```python
import pandas as pd
import matplotlib.pyplot as plt

#Load the data set
titanic = pd.read_csv("C:/Users/aparn/OneDrive/Documents/SJSU/DATA_230/Lectures/Data Source/Titanic-Dataset.csv")

# Average survival rate for each class
class_survival_rates = titanic.groupby('Pclass')['Survived'].mean()

print(class_survival_rates)

#Create Bar chart
plt.figure(figsize=(5, 4))
plt.bar(class_survival_rates.index, class_survival_rates, color=['red', 'green', 'blue'])
plt.xlabel("Passenger Class")
plt.ylabel("Average Survival Rate")
plt.title("Average Survival Rate by Passenger Class")
plt.show()
```

```python
#Check if Age and Fare has missing values

Missing_values_Age=titanic['Age'].isnull().sum()
Missing_values_Fare=titanic['Fare'].isnull().sum()

print(Missing_values_Age)
print(Missing_values_Fare)

#Median imputation to replace the missing values
if(Missing_values_Age)>0:
    titanic['Age'] = titanic['Age'].fillna(titanic['Age'].median())
if(Missing_values_Fare)>0:
    titanic['Fare'] = titanic['Fare'].fillna(titanic['Fare'].median())

plt.figure(figsize=(10, 6))

# Scatter plot for each passenger class
for pclass in titanic['Pclass'].unique():#iterates over each unique passenger class.
    class_data = titanic[titanic['Pclass'] == pclass]#extracts data specific to the current passenger class.
    plt.scatter(class_data['Age'], class_data['Fare'], label=f'Class {pclass}', alpha=0.5)

plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Relationship between Age and Fare by Passenger Class')
plt.legend()
plt.show()
```
