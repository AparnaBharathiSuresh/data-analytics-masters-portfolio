```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import warnings

# Load the data set
cars = pd.read_csv("C:/Users/aparn/OneDrive/Documents/SJSU/DATA_230/Lectures/Data Source/used_cars_data.csv")

# Print first few rows
print("First few rows:\n")
cars.head()
```

```python
# Print last few rows 
print("Last few rows:\n")
cars.tail()
```

```python
# Calculate percentage of missing values for each column
missing_percentage = (cars.isnull().mean() * 100).round(3)
print("\nPercentage of missing values for each column:\n")
print(missing_percentage)
```

```python
# Remove the "S.No." column
cars = cars.drop(columns=['S.No.'])

# Extract car brands from the "Name" column
cars['Brand'] = cars['Name'].str.split().str.get(0)

# Remove the original "Name" column
cars = cars.drop(columns=['Name'])

# Calculate car ages based on the "Year" column
current_year = 2024  
cars['Car_age'] = current_year - cars['Year']

# Remove the original "Year" column
cars = cars.drop(columns=['Year'])

# First few rows after editing the columns
print("First few rows after editing the columns:\n")
cars.head()
```

```python
# Last few rows after editing the columns
print("Last few rows after editing the columns:\n")
cars.tail()
```

```python
#Remove the units for the columns below and store only the numerical data 
def clean_and_convert(column):
    return pd.to_numeric(column.str.extract('(\d+\.?\d*)')[0], errors='coerce')

cars['Mileage'] = clean_and_convert(cars['Mileage'])
cars['Engine'] = clean_and_convert(cars['Engine'])
cars['Power'] = clean_and_convert(cars['Power'])
cars['New_Price'] = clean_and_convert(cars['New_Price'])


cars['Mileage']=cars['Mileage'].astype(float)
#cars['Engine'] = cars['Engine'].astype(float)
#cars['Power'] = cars['Power'].astype(float)
cars['New_Price'] = cars['Power'].astype(float)


# Generate histograms for numerical variables
numerical_columns = cars.select_dtypes(include='number').columns
for column in numerical_columns:
    sns.histplot(cars[column],kde=True,bins=20)
    plt.title(f'Histograms of {column}')
    plt.show()
```

```python
# Generate bar charts for categorical variables(example of object data type is string)
categorical_columns = cars.select_dtypes(include='object').columns
for column in categorical_columns:
    value_counts = cars[column].value_counts()
    ax = value_counts.plot(kind='bar', figsize=(12, 8), title=f'Bar chart for {column}',rot=45)
    ax.set_ylabel('Count')

    # Annotate values on top of the bars
    for i, value in enumerate(value_counts):
        plt.text(i, value + 0.1, str(value), ha='center', va='bottom')
    plt.show()
```

```python
# EDA Bivariate Analysis
# Create bar charts for average price versus categorical variables
for column in categorical_columns:
    #Calculate avg price
    avg_price_by_category = cars.groupby(column)['Price'].mean().round(3).sort_values(ascending=False)
    ax =avg_price_by_category.plot(kind='bar', figsize=(8, 6), title=f'Average Price vs {column}',rot=90)
    # Annotate values on top of the bars
    for i, value in enumerate(avg_price_by_category):
        if avg_price_by_category.notnull().all():
            plt.text(i, value + 0.1, str(value), ha='center', va='bottom',rotation=90)
    plt.ylabel('Average Price(Lakh Rupees)')
    plt.show()
```

```python
# EDA Multivariate Analysis

# Select only numeric columns for correlation matrix
numeric_columns = cars.select_dtypes(include='number').columns
correlation_matrix = cars[numeric_columns].corr()
#print(correlation_matrix)
# Generate a heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Correlation Matrix Heatmap")
plt.show()
```
