```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file
data=pd.read_csv("C:/Users/aparn/Downloads/cluster_data1.csv")

#Standardize the data
data_centered=data-data.mean(axis=0)

#SVD to decompose the data matrix
U, S, Vt = np.linalg.svd(data_centered)

#First two principal components
c1=Vt.T[:,0]
c2=Vt.T[:,1]

print("\n Principal component 1:\n")
print(c1)
print("\n Principal component 2:\n")
print(c2)

#Project the data onto the first two principal components
projected_data_c1 = np.dot(data_centered,c1)

projected_data_c2 = np.dot(data_centered,c2)

print("\nProjected_data_for_Principal_component_1:\n")
print(projected_data_c1)

print("\nProjected_data_for_Principal_component_2:\n")
print(projected_data_c2)
```

```python
# Plot the projected data in a 2D plot
plt.figure(figsize=(8, 6))
plt.scatter(projected_data_c1, projected_data_c2)
plt.xlabel("C1")
plt.ylabel("C2")
plt.title("Data Projected onto C1 and C2")
plt.grid(True)
plt.show()
```
