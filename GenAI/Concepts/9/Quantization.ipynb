```python
def average_inference_time(model, tokenizer, texts, label, max_new_tokens=30):
    latencies = []
    model.eval()
    device = next(model.parameters()).device
    tokenizer.padding_side = "left"
    tokenizer.pad_token_id = tokenizer.eos_token_id

    for text in texts[:100]:  # run on 100 samples
        inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True).to(device)
        torch.cuda.synchronize()
        start = torch.cuda.Event(enable_timing=True)
        end = torch.cuda.Event(enable_timing=True)
        start.record()
        with torch.no_grad():
            _ = model.generate(**inputs, max_new_tokens=max_new_tokens)
        end.record()
        torch.cuda.synchronize()
        latencies.append(start.elapsed_time(end) / 1000)  # sec

    avg = sum(latencies) / len(latencies)
    print(f"\n⏱️ {label} average inference time: {avg:.4f} sec")
    return avg
```

```python
def test_multiple_prompts(model, tokenizer, prompts, device="cuda", max_new_tokens=20, label="Model"):
    model.eval()
    model.to(device)

    total_time = 0.0
    results = []

    for idx, prompt in enumerate(prompts):
        print(f"🔍 Prompt {idx+1}: {prompt}")
        inputs = tokenizer(prompt, return_tensors="pt").to(device)

        # Warm-up (optional)
        _ = model.generate(**inputs, max_new_tokens=max_new_tokens)

        start_event = torch.cuda.Event(enable_timing=True)
        end_event = torch.cuda.Event(enable_timing=True)
        torch.cuda.synchronize()
        start_event.record()

        outputs = model.generate(
            **inputs,
            max_new_tokens=max_new_tokens,
            do_sample=True,
            temperature=0.9,
            top_k=50,
            repetition_penalty=1.1
        )

        end_event.record()
        torch.cuda.synchronize()
        inference_time = start_event.elapsed_time(end_event) / 1000  # seconds

        generated = tokenizer.decode(outputs[0], skip_special_tokens=True)
        print(f" {label} Output: {generated}")
        print(f" Inference Time: {inference_time:.4f} sec\n")

        results.append((prompt, generated, inference_time))
        total_time += inference_time

    avg_time = total_time / len(prompts)
    print(f"📊 Average Inference Time for {label}: {avg_time:.4f} sec")

    return results, avg_time
```

```python
def compute_perplexity(model, tokenizer, label):
    model.eval()
    device = next(model.parameters()).device
    data = load_dataset("wikitext", "wikitext-2-raw-v1", split="test[:1%]")
    text = "\n\n".join([d["text"] for d in data if d["text"].strip()])[:1500]
    inputs = tokenizer(text, return_tensors="pt", truncation=True, max_length=1024).to(device)
    input_ids = inputs["input_ids"]

    with torch.no_grad():
        outputs = model(input_ids=input_ids, labels=input_ids)
        loss = outputs.loss
        ppl = torch.exp(loss).item()

    print(f"📊 {label} Perplexity: {ppl:.2f}")
    return ppl

ppl_orig = compute_perplexity(model_orig, tokenizer_orig, "Original OPT-350m")
ppl_quant = compute_perplexity(model_quant, tokenizer_quant, "Quantized OPT-350m")
```

```python
import os

def get_model_size(path):
    total = 0
    for f in os.listdir(path):
        full_path = os.path.join(path, f)
        if os.path.isfile(full_path):
            total += os.path.getsize(full_path)
    return total / 1e6  # MB

size_orig = get_model_size("opt-350m-original")  # saved using .save_pretrained()
size_quant = get_model_size("opt-350m-GPTQ")
```

### Findings

This comparison evaluates the effectiveness of GPTQ 4-bit quantization (group size = 128) applied to two OPT models — facebook/opt-350m and facebook/opt-125m — using the Wikitext-2 dataset for both calibration and evaluation.

1. Inference Speed
Quantization consistently improved inference speed across both models.

OPT-125M saw the largest speed gain, improving from 1.07s → 0.83s (≈22% faster).

OPT-350M improved from 1.78s → 1.61s, showing ≈9% speedup.

2. Model Size Reduction
OPT-350M size reduced from 1328.28 MB → 219.85 MB (83% smaller).

OPT-125M reduced from 504.39 MB → 128.58 MB (75% smaller).

This enables deployment on memory-constrained devices while retaining functional performance.

3. Perplexity Impact
Perplexity increased slightly after quantization:

OPT-350M: 33.73 → 34.84

OPT-125M: 43.74 → 46.59

The performance degradation is minimal and acceptable for most downstream applications.

Conclusion
GPTQ 4-bit quantization provides an effective trade-off between model compression and performance. While both models benefited from reduced size and improved inference speed, OPT-350M demonstrated better language modeling ability with significantly lower perplexity overall. However, OPT-125M offers excellent efficiency, making it more suitable for real-time or edge deployment scenarios.
