```python
# Step 1: Import Libraries
import os
import openai
import pinecone
import torch
from datasets import load_dataset
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import pandas as pd

# Step 2: Set API Keys
os.environ["OPENAI_API_KEY"] = ""
os.environ["PINECONE_API_KEY"] = ""
openai.api_key = os.environ["OPENAI_API_KEY"]
```

```python
# Step 3: Load Dataset
dataset = load_dataset("jamescalam/ai-arxiv-chunked", split="train")
#dataset = dataset.filter(lambda x: x["chunk"] is not None)
#dataset = dataset.select(range(100))  # Use top 100 for demo
dataset
```

```python
dataset = dataset.map(lambda x: {
    "id": f'{x["id"]}-{x["chunk-id"]}',
    "text": x["chunk"],
    "metadata": {
        "title": x["title"],
        "url": x["source"],
        "primary_category": x["primary_category"],
        "published": x["published"],
        "updated": x["updated"],
        "text": x["chunk"],
    }
})
# drop uneeded columns
dataset = dataset.remove_columns([
    "title", "summary", "source",
    "authors", "categories", "comment",
    "journal_ref", "primary_category",
    "published", "updated", "references",
    "doi", "chunk-id",
    "chunk"
])
dataset
```

```python
from pinecone.grpc import PineconeGRPC

# get API key from app.pinecone.io
api_key = ""

# configure client
pc = PineconeGRPC(api_key=api_key)
```

```python
import time
from pinecone import ServerlessSpec

spec = ServerlessSpec(
    cloud='aws',        # or 'gcp'
    region='us-east-1'  # or whatever your region is
)

index_name = "rerankers"
existing_indexes = [
    index_info["name"] for index_info in pc.list_indexes()
]

# check if index already exists (it shouldn't if this is first time)
if index_name not in existing_indexes:
    # if does not exist, create index
    pc.create_index(
        index_name,
        dimension=768,  # dimensionality of e5-large
        metric='cosine',
        spec=spec
    )
    # wait for index to be initialized
    while not pc.describe_index(index_name).status['ready']:
        time.sleep(1)

# connect to index
index = pc.Index(index_name)
time.sleep(1)
# view index stats
index.describe_index_stats()
```

```python
from sentence_transformers import SentenceTransformer

device = "cuda" if torch.cuda.is_available() else "cpu"
model = SentenceTransformer("intfloat/e5-base").to(device)
```

```python
def embed(text_batch):
    # Add "passage: " prefix as required by E5 models
    prefixed = ["passage: " + text for text in text_batch]

    # Encode using GPU
    embeddings = model.encode(prefixed, batch_size=32, show_progress_bar=False, convert_to_numpy=True)
    return embeddings
```

```python
from pinecone import Pinecone, ServerlessSpec

#pc = Pinecone(api_key="your-api-key")

index_name = "arxiv-e5-index"

# delete if it already exists with wrong dimension
if index_name in pc.list_indexes().names():
    pc.delete_index(index_name)

# create with correct dimension = 768
pc.create_index(
    name=index_name,
    dimension=768,
    metric="cosine",
    spec=ServerlessSpec(cloud="aws", region="us-east-1")
)

index = pc.Index(index_name)
```

```python
from tqdm.auto import tqdm
import time
from concurrent.futures import ThreadPoolExecutor
import pinecone
from pinecone import PineconeException

batch_size = 96
executor = ThreadPoolExecutor(max_workers=1)

def safe_upsert(vectors):
    for attempt in range(3):
        try:
            index.upsert(vectors=vectors)
            return
        except PineconeException as e:
            print(f"⚠️ Pinecone error: {e}. Sleeping 10s (attempt {attempt+1})...")
            time.sleep(10)
        except Exception as e:
            print(f"❌ Unexpected error during upsert: {e}")
            break

# Main loop with parallel upload
for i in tqdm(range(0, len(dataset), batch_size)):
    i_end = min(len(dataset), i + batch_size)
    batch = dataset[i:i_end]

    # ✅ Embed using GPU (local E5 model)
    text_batch = ["passage: " + text for text in batch["text"]]
    embeds = model.encode(
        text_batch,
        batch_size=32,
        show_progress_bar=False,
        convert_to_numpy=True
    )

    to_upsert = list(zip(batch["id"], embeds.tolist(), batch["metadata"]))

    # ✅ Offload the upload to another thread
    executor.submit(safe_upsert, to_upsert)
```

```python
def get_docs(query: str, top_k: int = 5) -> list[dict]:
    # ✅ Add E5 query prefix
    query_input = ["query: " + query]

    # ✅ Local GPU-based query embedding
    xq = model.encode(query_input, convert_to_numpy=True)[0]

    # ✅ Query Pinecone
    res = index.query(vector=xq.tolist(), top_k=top_k, include_metadata=True)

    # ✅ Format results
    docs = [{
        "id": str(i),
        "text": match["metadata"]["text"]
    } for i, match in enumerate(res["matches"])]

    return docs
```

```python
docs = get_docs("What is retrieval augmented generation?", top_k=5)
for doc in docs:
    print(f"🔍 ID: {doc['id']}\n📄 {doc['text'][:300]}...\n")
```

```python
rerank_name = "bge-reranker-v2-m3"

def compare(query: str, top_k: int, top_n: int):
    # Step 1: Get top_k documents from vector search
    docs = get_docs(query, top_k=top_k)

    # Step 2: Use Pinecone hosted reranking
    reranked = pc.inference.rerank(
        model=rerank_name,
        query=query,
        documents=docs,
        top_n=top_n,
        return_documents=True
    )

    # Step 3: Compare original and reranked results
    original_docs = []
    reranked_docs = []

    print("[ORIGINAL] -> [RERANKED]")
    for i, doc in enumerate(reranked.data):
        print(f"{doc.index} -> {i}")
        if i != doc.index:
            reranked_docs.append(f"[{doc.index}]\n{doc['document']['text']}")
            original_docs.append(f"[{i}]\n{docs[i]['text']}")
        else:
            reranked_docs.append(doc["document"]["text"])
            original_docs.append(None)

    for orig, rerank in zip(original_docs, reranked_docs):
        if not orig:
            print(f"SAME:\n{rerank}\n\n---\n")
        else:
            print(f"ORIGINAL:\n{orig}\n\nRERANKED:\n{rerank}\n\n---\n")
```

```python
query = "What is retrieval augmented generation?"
compare(query, top_k=25, top_n=5)
```

## Conclusion and Documentation:

### First-Stage Retrieval (Vector Similarity)
- Used intfloat/e5-base to embed both documents and query.
- Queried top 25 documents from the Pinecone vector index using cosine similarity.
- The results included many relevant papers about NLP, RAG, hallucination, and generative models.
- However, some of the top-ranked documents focused broadly on citations, summaries, or general-purpose systems without deep contextual relevance to the query intent.

### Second-Stage Retrieval (Pinecone Reranker: bge-reranker-v2-m3)
- Applied Pinecone's hosted reranker on the top 25 vector-retrieved documents.
- The reranker reordered documents based on cross-encoder relevance to the input query.
- Top reranked documents focused much more precisely on:
  - Knowledge hallucination
  - Application of RAG to open-domain dialogue
  - Training methods for improving RAG performance
- Example: A document originally ranked 4th was reranked to 1st because it directly addressed the hallucination reduction problem in RAG-based systems.

### Final Conclusion

Two-stage retrieval dramatically improved the relevance and clarity of results:
-  First-stage ensured semantic coverage using dense embeddings.
-  Second-stage filtered and elevated the most informative and contextual documents using Pinecone's reranker.
-  It reduced noise from generic or loosely related content.
-  Ideal for tasks like question answering, document grounding, and hallucination mitigation.
